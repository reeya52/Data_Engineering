```python
import pandas as pd
import numpy as np

df_asc_data = pd.read_csv('acs2017_census_tract_data.csv', usecols = ['County', 'State', 'TotalPop', 'Poverty', 'IncomePerCap'])

group_by = df_asc_data.groupby(['County', 'State']).groups
# print(group_by)
county_info = pd.DataFrame(columns = ['Id', 'County', 'State', 'TotalPop', 'Poverty', 'IncomePerCap'])
id = 0
for group in group_by:
#     print(group)
    total_population = 0
    total_poverty = 0
    total_income = 0
    for i in group_by[group]:
        row = df_asc_data.loc[i]
#         print(row)
        if np.isnan(row['Poverty']) or np.isnan(row['IncomePerCap']):
            continue
        population = row['TotalPop']
        total_population += population
        total_poverty += (row['Poverty'] if row['Poverty'] else 0) * population / 100
        total_income += (row['IncomePerCap'] if row['IncomePerCap'] else 0)
    current_group = df_asc_data.loc[group_by[group][0]]
    percent_poverty = total_poverty * 100 / total_population
    income_percapita = total_income / total_population
    county_info.loc[id] = [id+1, current_group['County'], current_group['State'], total_population, percent_poverty, income_percapita]
    id += 1

county_info
```

```python
df_covid_data = pd.read_csv('COVID_county_data.csv')

df_covid_data['year_month'] = pd.to_datetime(df_covid_data['date']).dt.to_period('M')
df_covid_data
```

```python
group_by = df_covid_data.groupby(['county', 'state', 'year_month']).groups
print(group_by)
```

```python
covid_info = pd.DataFrame(columns=['Id', 'Date', 'County', 'State', 'Cases', 'Deaths'])
id = 0
for group in group_by:
    total_cases = 0
    total_deaths = 0
    for i in group_by[group]:
        row = df_covid_data.loc[i]
        total_cases += row['cases']
        total_deaths += row['deaths']
    covid_info.merge(county_info, left_on='Id', right_on='Id')
    covid_info.loc[id] = ['Id', group[2], group[0], group[1], total_cases, total_deaths]
    id += 1
    
covid_info
```
