```python
import pandas as pd

df = pd.read_csv('books.csv')
```

```python
# Filtering using drop method
df2 = df.drop(['Edition Statement','Corporate Author', 'Corporate Contributors', 'Former owner', 'Engraver', 'Issuance type', 'Shelfmarks'], axis=1)
df2
```

```python
# Tidying Up the Data
import re

pattern = '^.*(\d{4})[^?]*$'
df2['Date of Publication'] = df2['Date of Publication'].str.extract(pat = pattern)
df2
```

```python
# Tidying with applymap()

df_uniplaces = open('uniplaces.txt', "r", encoding='utf-8')
lines = df_uniplaces.readlines()
df_uniplaces.close()

state_list = []
city_list = []
uni_list = []

for line in lines:
    txt = line.strip()
    x = re.findall("\[edit]\Z", txt)
    y = re.findall(r'\(.*?\)', txt)

    if x:
        state = re.sub("\[edit]\Z", "", txt)
    
    if y:
        city = re.sub(r'\(.*?\)', "", txt)
        state_list.append(state)
        city_list.append(city)
        uni_list.append(y)

df_divided = pd.DataFrame({'State': state_list, 'City': city_list, 'Universities': uni_list})
# df_divided

def tidy_university_places(item):
    retval = re.sub(r'\[\d*\]', '', str(item))
    return retval

df1_divided = df_divided.applymap(tidy_university_places)
df1_divided
```
